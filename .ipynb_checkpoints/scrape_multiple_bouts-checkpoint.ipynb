## Beautiful Soup setup

```python
# ----------------- Beautiful Soup Setup ------------------ #

# all yokozuna
# db_url = "http://sumodb.sumogames.de/Rikishi.aspx?shikona=&heya=-1&shusshin=-1&b=-1&high=1&hd=-1&entry=-1&intai=-1&sort=4"

# three current yokozuna
db_url = "http://sumodb.sumogames.de/Rikishi.aspx?shikona=&heya=-1&shusshin=-1&b=201701&high=1&hd=-1&entry=-1&intai=-1&sort=4"

# all makushita
# db_url = "http://sumodb.sumogames.de/Rikishi.aspx?shikona=&heya=-1&shusshin=-1&b=-1&high=7&hd=-1&entry=-1&intai=-1&sort=4"

# all rikishi
# db_url = "http://sumodb.sumogames.de/Rikishi.aspx?shikona=&heya=-1&shusshin=-1&b=-1&high=-1&hd=-1&entry=-1&intai=-1&sort=7"

source = urllib2.urlopen(db_url).read()
soup = bs4.BeautifulSoup(source, 'lxml')  # turn into soup

print 'done'
```

## Obtain Rows from Database & Initialize DataFrame

```python
# obtain rows in database (rikishi to loop over)
right_td = soup.findAll('td', {'class': 'layoutright'})
rows = right_td[0].findAll('tr')  # take first td with layoutright class

# initialize dataframe
cols = [u'ID1', u'shikona1', u'ID2', u'shikona2', u'ID1_wins', u'ID2_wins', u'flag']
h2h_df = pd.DataFrame(columns=cols)
```

# Loop Through Multiple Profiles

```python
start = time.time()

logical = rikishi_df['shikona'].map(lambda x: x is not None) & \
          rikishi_df['sumo_stable'].map(lambda x: x is not None) & \
          rikishi_df['bday'].map(lambda x: pd.notnull(x)) & \
          rikishi_df['debut'].map(lambda x: pd.notnull(x))
filtered_df = rikishi_df[logical]

# iterate over each rikishi
for ind, row in enumerate(rows):
    if ind:  # skip first headers row
        
        # --- Obtain Current Rikishi URL --- #
        links = row.findAll('a', href=True)
        extension = links[0]['href']  # access link inside href
        rikishi_url = "http://sumodb.sumogames.de/" + extension  # rikishi1 basic profile
        # extract shikona & bout url (need to write a function to do this simultaneously)
        
        # --- Obtain Shikona, ID, and Bout URL --- #
        shikona1, bout_url = bouturl_scrape(rikishi_url)
        ID1 = rikishi_df.loc[rikishi_df['shikona'] == shikona1, 'rikishi_ID'].item()
        print "\n# --- Iter: %d, " % ind + shikona1 + " --- #\n"
        
        # --- Scrape H2H Bout Data --- #
        data_rows = h2h_scrape(shikona1, ID1, bout_url, h2h_df, filtered_df)
        
        # --- Store Data in Overall Dataframe --- #
        if data_rows:  # add rows only if data_rows is full
            h2h_df = h2h_df.append(data_rows, ignore_index=True)

h2h_df = h2h_df[cols] # reorganize columns
        
end = time.time()
exec_time = end - start
print "SCRAPING DONE"
print "Execution Time: %.2f (s)" % exec_time
```

## Check For Duplicate Rikishi Profiles

```python
logical = rikishi_df['shikona'].map(lambda x: x is not None) & \
          rikishi_df['sumo_stable'].map(lambda x: x is not None) & \
          rikishi_df['bday'].map(lambda x: pd.notnull(x)) & \
          rikishi_df['debut'].map(lambda x: pd.notnull(x))
filtered_df = rikishi_df[logical]

shikona_list = []
heya_list = []
for _, row in filtered_df.iterrows():
    shikona = row['shikona']
    heya = row['sumo_stable']
    if shikona not in shikona_list:
        logical = (filtered_df['shikona'] == shikona) & (filtered_df['sumo_stable'] == heya)
        if len(filtered_df[logical]) > 1:
            shikona_list.append(shikona)
            heya_list.append(heya)
            
print len(shikona_list)
# at least 551 sumo profiles have duplicate shikona names
# need to handle these cases
for row in zip(shikona_list, heya_list):
    print row

# rikishi with useful records (more than just a few 1-0/0-1):
# Yamaguchi, Tokitsukaze (only 1)
```
